```python
import matplotlib.pyplot as plt

# Define the original and improved color pairs
color_A = ['dodgerblue',    'green',    'orangered',    'indigo',   'sienna']
color_B = ['darkorange',    'orchid',   'turquoise',    'gray',     'limegreen']

# Function to plot color pairs
def plot_color_pairs(color_A, color_B, title, position):
    for i, (a, b) in enumerate(zip(color_A, color_B)):
        plt.plot([0, 1], [i, i], color=a, linewidth=20, label=f"A{i+1}" if position == 1 else "")
        plt.plot([1, 2], [i, i], color=b, linewidth=20, label=f"B{i+1}" if position == 1 else "")
    plt.yticks(range(len(color_A)), [f"Pair {i+1}" for i in range(len(color_A))])
    plt.xticks([])
    plt.title(title, fontsize=14)

# Create the plot
plt.figure(figsize=(10, 6))
# Plot original pairs
plt.subplot(1, 2, 1)
plot_color_pairs(color_A, color_B, "Original Color Pairs", position=1)
plt.tight_layout()
plt.show()
```
